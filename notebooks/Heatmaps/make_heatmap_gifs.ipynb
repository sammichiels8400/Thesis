# Make GIFS

This notebook will make GIFS of the different pictures so we can easily see the differences.


```python
import os
import imageio

def is_image(file_name, root):
    """Check if the file is an image based on its extension."""
#     if file_name.lower().endswith('.gif'):
#         os.remove(os.path.join(root, file_name))
    image_extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.tiff')
    return file_name.lower().endswith(image_extensions)

def create_gif_from_images(image_files, folder_path):
    """Create a GIF from the given image files."""
    images = []
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        images.append(imageio.imread(image_path))
    gif_path = os.path.join(folder_path, 'output.gif')
    if len(images) > 0:
        imageio.mimsave(gif_path, images, duration=0.5)  # Duration can be adjusted
        print(f"GIF created and saved at: {gif_path}")

def traverse_folders(folder_path):
    """Traverse folders recursively and create GIFs in folders with only images."""
    for root, dirs, files in os.walk(folder_path):
        # Check if the current folder contains only image files
        image_files = [f for f in files if is_image(f, root)]
        
        if len(image_files) == len(files):  # All files are images
            create_gif_from_images(image_files, root)


# Replace this path with the path to the folder you want to process
folder_path = "../../heatmaps"
traverse_folders(folder_path)
```
