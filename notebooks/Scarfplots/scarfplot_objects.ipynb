```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt


pad = "../../data/csv/"
image_pad =  "../../data/Images/CrimeScenes/"

df = pd.read_csv(pad + 'fix_evidence.csv')

df['Participant_experience'] = df['Participant'].astype(str) + '_' + df['experience'].astype(str)

df_filtered = df[(df['Image'] == 2) & (df['Scene'] == 3)]
```

# Control


```python
df_experience = df_filtered[df_filtered['experience'] == "Control"].copy()

df_experience['Name'] = df_experience['Name'].apply(
    lambda x: x if x in ['Body', 'Blood_on_floor', 'Blood_on_table', 'Knife_on_floor', 'Knife_on_plate'] else 'Other'
)
df_experience = df_experience.dropna(subset=['Name'])

# Sort and compute cumulative fixation start times per participant
df_experience.sort_values(['Participant_experience', 'FixNr'], inplace=True)
df_experience['CumulativeStart'] = (
    df_experience.groupby('Participant_experience')['FixDur'].cumsum() - df_experience['FixDur']
)

# Assign y-position per participant
participants = df_experience['Participant_experience'].unique()
y_positions = {p: i for i, p in enumerate(participants)}

# Define colors
colors = {
    'Blood_on_floor': 'red',
    'Blood_on_table': 'green',
    'Body': 'cyan',
    'Knife_on_floor': 'black',
    'Knife_on_plate': 'yellow',
    'Other': 'blue'
}
# Plot
fig, ax = plt.subplots(figsize=(10, 5))

for _, row in df_experience.iterrows():
    ax.broken_barh(
        [(row['CumulativeStart'], row['FixDur'])],
        (y_positions[row['Participant_experience']] - 0.4, 0.8),
        facecolors=colors[row['Name']]
    )

# Formatting
ax.set_yticks(range(len(participants)))
ax.set_yticklabels(participants)
ax.set_xlabel("Cumulative Fixation Time")
ax.set_title("Scarf Plot of Fixations on Objects (Control)")
ax.grid(True, linestyle="--", alpha=0.5)
ax.set_xlim([0, 80000])

plt.show()
```
